```python
df_features = pd.DataFrame(bc_df.data , columns=bc_df.feature_names)
df_target   = pd.DataFrame(bc_df.target,columns=["cancer"])
```

```python
# join the two dfs
```

```python
display(df_features.head(2),df_target.head(2))
```

```python
df = pd.concat([df_features,df_target] ,axis= 1)
```
